```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df
```

```python
#KMeans - Clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
scaled_df['K_3'] = kmeans.fit_predict(scaled_df)
scaled_df
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(scaled_df, scaled_df['K_3'])
score
```

```python
#Elbow method to find the best number of clusters
scores = []

for i in range(2,11):
    cluster = KMeans(n_clusters=i)
    scaled_df["K_3"] = cluster.fit_predict(scaled_df)
    scores.append(silhouette_score(scaled_df,scaled_df["K_3"]))

scores


import matplotlib.pyplot as plt
plt.scatter(range(2,11), scores)
plt.plot(range(2,11), scores,color='red')
plt.show()
```

```python
#Reloading the dataset to perform the Hierarchial Clustering
import pandas as pd
df = pd.read_excel('D:\\Q&A PDFS\\EastWestAirlines.xlsx', sheet_name='data')
df
```

```python
df = df.drop(columns=['ID#', 'Award?'])
df
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df.reset_index(drop=True, inplace=True)
scaled_df
```

```python
#Hierarchial Clustering
from sklearn.cluster import AgglomerativeClustering
cluster1 = AgglomerativeClustering(n_clusters=5,linkage='single')
scaled_df["Single"] = cluster1.fit_predict(scaled_df) #Single linkage clustering

cluster2 = AgglomerativeClustering(n_clusters=5,linkage='complete')
scaled_df["complete"] = cluster2.fit_predict(scaled_df) #complete linkage clustering

cluster3 = AgglomerativeClustering(n_clusters=5,linkage='average')
scaled_df["average"] = cluster3.fit_predict(scaled_df) #average linkage clustering

cluster4 = AgglomerativeClustering(n_clusters=5,linkage='ward')
scaled_df["ward"] = cluster4.fit_predict(scaled_df) #ward linkage clustering

scaled_df
```

```python
# cluster formation -single-
import numpy as np
import matplotlib.pyplot as plt
plt.scatter(scaled_df['Balance'], scaled_df['Bonus_trans'], c=scaled_df["Single"], cmap='rainbow')  
plt.show()

from sklearn.metrics import silhouette_score
score=silhouette_score(scaled_df,scaled_df["Single"])
print("single linkage score:", np.round(score,3))
```

```python
# cluster formation -complete-
import matplotlib.pyplot as plt
plt.scatter(scaled_df['Balance'], scaled_df['Bonus_trans'], c=scaled_df["complete"], cmap='rainbow')  
plt.show()

from sklearn.metrics import silhouette_score
score=silhouette_score(scaled_df,scaled_df["complete"])
print("complete linkage score:", np.round(score,3))
```

```python
# cluster formation -average- good
import matplotlib.pyplot as plt
plt.scatter(scaled_df['Balance'], scaled_df['Bonus_trans'], c=scaled_df["average"], cmap='rainbow')  
plt.show()

from sklearn.metrics import silhouette_score
score=silhouette_score(scaled_df,scaled_df["average"])
print("average linkage score:", np.round(score,3))
```

```python
# cluster formation -ward
import matplotlib.pyplot as plt
plt.scatter(scaled_df['Balance'], scaled_df['Bonus_trans'], c=scaled_df["ward"], cmap='rainbow')  
plt.show()

from sklearn.metrics import silhouette_score
score=silhouette_score(scaled_df,scaled_df["ward"])
print("ward linkage score:", np.round(score,3))
```

```python
# Analyze Hierarchial Clustering
# The Agglomerative clustering with different linkages shows varying results.
# Single linkage tends to form elongated clusters, which might not be ideal for this dataset based on the scatter plots. The silhouette score is low.
# Complete linkage provides a slightly better structure than single linkage based on the scatter plot, but the silhouette score is still not very high.
# Average linkage appears to produce more balanced clusters compared to single and complete linkage based on the scatter plot. The silhouette score is better than single and complete.
# Ward linkage seems to create more compact clusters and has the highest silhouette score among the Agglomerative methods tested, suggesting it might be the best fit for this dataset among the hierarchical methods explored.
```

```python
import pandas as pd
df = pd.read_excel('D:\\Q&A PDFS\\EastWestAirlines.xlsx', sheet_name='data')
df
df = df.drop(columns=['ID#', 'Award?'])
df
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df.reset_index(drop=True, inplace=True)
scaled_df
```

```python
from sklearn.cluster import DBSCAN
cluster = DBSCAN(eps=5,min_samples=20)
scaled_df["dbscan"] = cluster.fit_predict(scaled_df)

scaled_df.head()
```

```python
scaled_df["dbscan"].value_counts()
```

```python
scaled_df[scaled_df["dbscan"] == -1]

df_new = scaled_df[scaled_df["dbscan"] != -1]
df_new.shape
```

```python
df_new.reset_index(drop=True, inplace=True)
df_new
```

```python
score=silhouette_score(df_new,df_new["dbscan"])
print("dbscan linkage score:", np.round(score,3))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(scaled_df['Balance'], scaled_df['Bonus_trans'], c=scaled_df['dbscan'], cmap='viridis', marker='o', s=50)
plt.title('DBSCAN Clustering Results')
plt.xlabel('Balance')
plt.ylabel('Bonus_trans')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
# Analyze DBSCAN Clustering
# DBSCAN with the chosen parameters identified a significant number of noise points (-1).
# The remaining points are grouped into two clusters (0 and 1).
# Cluster 0 seems to represent the majority of the data points, likely the 'typical' customers.
# Cluster 1 appears to be a smaller cluster, potentially representing a group with different characteristics, possibly related to outlier-like behavior in some features that the epsilon parameter allowed them to cluster together.
# The silhouette score for DBSCAN is calculated on the non-noise points and is quite good, indicating well-separated clusters among the identified groups, excluding the outliers.
# The scatter plot for DBSCAN visually highlights the two main clusters and the scattered noise points.
```

```python
# Overall Insights:
# K-Means (with k=9) and Agglomerative Clustering (Ward linkage) seem to provide potentially meaningful groupings of customers based on the features.
# DBSCAN identifies potential outliers and separates them, which can be valuable for understanding unusual customer behavior.
# The choice of the best clustering algorithm and the optimal number of clusters depends on the specific business goals and the interpretation of the cluster characteristics.
# To get a comprehensive understanding of each cluster's profile for all algorithms, it's essential to calculate the mean (or median) values of all original features within each cluster and interpret these values in the context of the airline's data.
```
