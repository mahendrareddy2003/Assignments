```python
#Standardizing the features to ensure they have a mean of 0 and a standard deviation of 1.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns = df.columns, index = df.index)
scaled_df
```

```python
#Implementing the PCA to reduce the dimensionality of the dataset
from sklearn.decomposition import PCA
pca = PCA()

PC = pca.fit_transform(scaled_df)

PC = pd.DataFrame(PC)
PC.head()
```

```python
#Determining the optimal number of the principal components using the technique called cumulative explained variance
```

```python
import matplotlib.pyplot as plt
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()
```

```python
#as per the above plot we can see the 90% of the data at the 7th principal component so i determine that the optimal number of components to be taken as 
# 7
```

```python
#Applying the PCA to the scaled_df with the number of components being 7
pca = PCA(n_components=7)
PC = pca.fit_transform(scaled_df)
PC = pd.DataFrame(PC)
PC.head()
```

```python
#Applying the KMeans clustering method with K-value = 3
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
scaled_df['K_3'] = kmeans.fit_predict(scaled_df)
scaled_df
```

```python
#visualization of clusters with the columns of the original scaled_df
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(scaled_df['Alcohol'], scaled_df['Malic'], c=scaled_df['K_3'], cmap='viridis')
plt.title('KMeans Clustering Results on Scaled Data')
plt.xlabel('Alchol')
plt.ylabel('Malic')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
# again performing the KMeans clustering method with same numbers clusters K=3 but this time with the PCA application
kmeans = KMeans(n_clusters=3)
PC['K_3'] = kmeans.fit_predict(PC)
PC
```

```python
#visualization of clusters with the columns of the PCA applied dataset PC 

plt.figure(figsize=(10, 6))
plt.scatter(PC[0], PC[1], c=PC['K_3'], cmap='viridis')
plt.title('KMeans Clustering Results on PC')
plt.xlabel('PC label 0')
plt.ylabel('PC label 1')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
#calculating the accuracy scores by using the silhouette score
from sklearn.metrics import silhouette_score
scaled_df_score = silhouette_score(scaled_df, scaled_df['K_3'])
print('silhouette score of scaled original dataset', scaled_df_score)
PC_score = silhouette_score(PC, PC['K_3'])
print('silhouette score of applied PCA dataset', PC_score)
```

```python
##Comparing the clustering results obtained from the sclaed original dataset and PCA-transformed data

#We can observe that the silhouette score of the scaled original dataset is 0.33, while the silhouette score of the dataset after applying PCA is 0.389. 
#which means that PCA transformed data silhouette score is higher than the silhouette score of the scaled original dataset.
#therfore we can say that applying the PCA improved the clustering results
```

```python
##trade-offs between clustering on the original scaled dataset versus the PCA-transformed dataset:

#clustering on original scaled dataset
#pros:
#no loss of the of the data.
#original features stays same.
#cons:
#High dimensional dataset.
#clustering performance can be lacking a bit.
#visualizing high dimensional dataset will be difficult

#clustering on PCA- transformed dataset
#pros:
#Reduces dimensionality.
#improves clustering performance.
#Easy to visualize the clustering results
#cons:
#Some information loss due to dimensionality reduction.
#we will lose the terms(column names) of the original features.
```

```python
##key findings and insights

#The scatter plot on scaled original data shows less distinct clusters, with more overlap between them.
#The scatter plot on the PCA-transformed data shows more clearly separated clusters, which aligns with the higher silhouette score.
#Therefore, applying PCA improved the clustering results with more distinct clusters
```

```python
##practical implications of using PCA and clustering in data analysis

#clustering performance improved when combined with the PCA
#we can handle the high dimensional dataset with the combination of PCA and clustering in data analysis.
#meaningful clustering patterns like we observed in the above plots
```

```python
##recommendations for when to use each technique based on the analysis conducted.

# When the dataset has less number of features it is recommended to use clustering with the original dataset without applying PCA.
# When preserving all the original information is important them also it is preferred to use clustering with the original dataset without applying PCA.

# when the dataset has more number of features it is recommended to use clustering with applied PCA dataset.
# When preserving all the original information is not that much important because of the large enough data then it is better to use clustering with-
# -PCA applied dataset.
```
