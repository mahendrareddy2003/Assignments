```python
#Handling missing values
df_train['Age'].fillna(df_train['Age'].median(), inplace=True)
```

```python
df_test['Age'].fillna(df_test['Age'].median(), inplace=True)
```

```python
df_train['Embarked'].fillna(df_train['Embarked'].mode()[0], inplace=True)
```

```python
df_test['Fare'].fillna(df_test['Fare'].median(), inplace=True)
```

```python
df_train.isnull().sum()
```

```python
#removing variable "cabin" from both data sets because of too many null values
df_train = df_train.drop('Cabin', axis=1)
df_test = df_test.drop('Cabin', axis=1)
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
#encoding categorical variables
df_train = pd.get_dummies(df_train, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
df_test = pd.get_dummies(df_test, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
df_train = pd.get_dummies(df_train, columns=['Pclass'], prefix='Pclass', drop_first=True)
```

```python
df_test = pd.get_dummies(df_test, columns=['Pclass'], prefix='Pclass', drop_first=True)
```

```python
#building a Logistic Regression model
from sklearn.linear_model import LogisticRegression
features = ['Age', 'Fare', 'SibSp', 'Parch', 'Sex_male', 'Embarked_Q', 'Embarked_S', 'Pclass_2', 'Pclass_3']
X_train = df_train[features]
Y_train = df_train['Survived']
X_test = df_test[features]
model = LogisticRegression(random_state=42, solver='liblinear') # Using liblinear solver for smaller datasets
model.fit(X_train, Y_train)
```

```python
#making predictions using testing data(testing data features)
Y_test_pred = model.predict(X_test)
```

```python
# Since we don't have the true labels for the test set ('Survived' column is missing),
# we cannot evaluate the model performance on df_test directly using standard metrics.
# The performance metrics (accuracy, precision, recall, F1, ROC-AUC) require-
# -the true labels of the test set to be compared with the predicted labels.
```

```python
# Since the provided df_test does not have the 'Survived' column,
# we cannot generate the true y_test for evaluating the model performance on it.
# To visualize the ROC curve, we need true labels.
# We will use the training data for demonstration purposes to plot the ROC curve.
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score
```

```python
Y_train_pred_proba = model.predict_proba(X_train)[:, 1]
fpr, tpr, thresholds = roc_curve(Y_train, Y_train_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc_score(Y_train, Y_train_pred_proba):.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

print(f"ROC-AUC Score on Training Data: {roc_auc_score(Y_train, Y_train_pred_proba):.2f}")
```

```python
#interpretting the coefficients
print("\nLogistic Regression Model Coefficients:")
coefficients = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': model.coef_[0]})
print(coefficients)
```

```python
# Significance of Features in Predicting Survival Probability:
# Based on the Logistic Regression model coefficients, the most influential features for predicting survival are:
# **Sex_male:** A large negative coefficient indicates being male significantly decreases the probability of survival.
# **Pclass (Pclass_2, Pclass_3):** Negative coefficients for Pclass 2 and 3 show lower survival probability compared to Pclass 1
# **Fare:** A positive coefficient suggests a higher fare paid increases survival probability (likely correlated with Pclass)
# **Age:** A generally negative coefficient implies older age is associated with a slightly lower survival probability
# **SibSp and Parch:** These features tend to have less significant impacts compared to Sex and Pclass.
# These insights highlight the critical roles of gender and socioeconomic status (represented by Pclass and Fare) in determining survival outcomes during the Titanic disaster.
```
