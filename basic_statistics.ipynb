```python
for col in numerical_col.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1
    outliers = df[(df[col]<Q1-1.5*IQR) | (df[col]>Q3+1.5*IQR)]
    print(f"{col}:{len(outliers)} outliers")
```

```python
# inferences:
#we see max.outliers in avg price 
#most of the variables shows right skewness
#except for the discount rate %
#almost all the variables have the outliers within them
```

```python
df_standardized = df.copy()
for col in numerical_col:
    mean=df[col].mean()
    std=df[col].std()
    df_standardized[col] = (df[col]-mean)/std
df_standardized.hist(figsize=(15,10))
```

```python
df_standardized[categorical_cols]
```

```python
# get-dummies
# df_encoded =pd.get_dummies(df_standardized, columns=categorical_cols, drop_first='false', dtype='int')
# df_encoded
```

```python
#one hot encoding
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder(sparse_output = False)
encoded = OHE.fit_transform(df_standardized[categorical_cols])
encoded_df = pd.DataFrame(encoded, columns=OHE.get_feature_names_out(categorical_cols))
final_df = pd.concat([df_standardized.drop(columns=categorical_cols), encoded_df], axis=1)
```
